MLP FOR CLASSIFICATION WITH PROPER CLASS OUTPUT



```python
import pandas as pd
df=pd.read_csv('/content/house-prices.csv')
df.head(5)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
df.info()
```

```python
bins=[0,100000,300000,500000,np.inf]
labels=['low','medium','high','very high']
df['price_category']=pd.cut(df['Price'],bins=bins,labels=labels)
```

```python
df.head(5)
```

```python
label_encoder=LabelEncoder()
df['price_category']=label_encoder.fit_transform(df['price_category'])
df.head(5)
```

```python
x=df.drop(['Price','price_category'],axis=1)
y=df['price_category']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train=pd.get_dummies(x_train,drop_first=True)
x_test=pd.get_dummies(x_test,drop_first=True)
x_train,x_test=x_train.align(x_test,join='outer',axis=1,fill_value=0)
```

```python
x_train.fillna(x_train.mean(),inplace=True)
x_test.fillna(x_test.mean(),inplace=True)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
def create_model(input_dim):
  model=Sequential([
      Dense(128,activation='relu',input_dim=input_dim),
      Dropout(0.3),
      Dense(64,activation='relu'),
      Dropout(0.2),
      Dense(32,activation='relu'),
      Dense(4,activation='softmax')])
  model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
  return model
```

```python
mlp_model=create_model(x_train.shape[1])
history=mlp_model.fit(x_train,y_train,epochs=50,batch_size=16)
```

```python
def fine_tune_model(model,x_train,y_train,x_test,y_test,epochs=20):
  model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=epochs,batch_size=16)
  return model
```

```python
mlp_model=fine_tune_model(mlp_model,x_train,y_train,x_test,y_test)
```

```python
test_loss,test_acc=mlp_model.evaluate(x_test,y_test)
print(f'Test Accuracy:{test_acc*100}.2f')
```
