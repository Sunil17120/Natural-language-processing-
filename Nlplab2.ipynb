```python
nlp=spacy.load("en_core_web_sm")
doc=nlp(data)
print("text\tlemma\tpos\ttag\tdep\tshape_")
for token in doc:
  print(token.text,"\t",token.lemma_,"\t",token.pos_,"\t",token.tag_,"\t",token.dep_,"\t",token.shape_)
```

```python
outputfile="output.txt"
f=open("output.txt","a")
outputstr="text\ttokken.lemma\tpos\ttag\tshape_\n"
f.write(outputstr)
for token in doc:
  outputstr=token.text+"\t"+token.lemma_+"\t"+token.pos_+"\t"+token.tag_+"\t"+token.dep_+"\t"+token.shape_+"\n"
  f.write(outputstr)
f.close()
```
